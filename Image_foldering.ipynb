```
import warnings
warnings.filterwarnings("ignore")
import os
import cv2
import pickle
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import hog
from google.colab.patches import cv2_imshow 
from skimage.filters import gabor
from skimage.transform import resize
from os import listdir
from os.path import isfile, join
import shutil
%matplotlib inline
```

```
def Image_Foldering(loaded_model,Current_folder,List_of_destination):
      
    for filename in os.listdir(Current_folder):

          print(filename)
          image_path = Current_folder + '/' + filename
          
          img = cv2.imread(image_path)

          #Image thresholding
          kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
          erosion = cv2.erode(img, kernel, iterations = 1)
          gray = cv2.cvtColor(erosion, cv2.COLOR_BGR2GRAY)
          blur = cv2.GaussianBlur(gray, (3,3), 0)
          thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

          #Resize image
          img_array =np.array(thresh)
          img_resized = resize(img_array,(28,28))
          
          #HOG FE
          x, hog_img_prd = hog(img_resized, orientations=9, pixels_per_cell=(4,4), block_norm='L2', feature_vector= True,
                              cells_per_block=(2, 2), visualize=True, multichannel=False)
          
          #Prediction
          pred = loaded_model.predict(x.reshape(1, -1))
          print(pred)


          if pred == 0:
            shutil.move(image_path,List_of_destination[0] )
          elif pred == 1:
            shutil.move(image_path,List_of_destination[1] ) 
          elif pred == 2:
            shutil.move(image_path,List_of_destination[2] )
          elif pred == 3:
            shutil.move(image_path,List_of_destination[3] ) 
          elif pred == 4:
            shutil.move(image_path,List_of_destination[4] )
          elif pred == 5:
            shutil.move(image_path,List_of_destination[5] ) 
          elif pred == 6:
            shutil.move(image_path,List_of_destination[6] )
          elif pred == 7:
            shutil.move(image_path,List_of_destination[7] )   
          elif pred == 8:
            shutil.move(image_path,List_of_destination[8] )
          elif pred == 9:
            shutil.move(image_path,List_of_destination[9] ) 

    
```

```
def move_files_from_one_folder_to_another(current_folder,destination): 
  
  files = os.listdir(current_folder)
  
  for f in files:
    x = current_folder +'/' + f
    shutil.move(x,destination )
```
